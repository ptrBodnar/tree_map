```python
import pandas as pd
import numpy as np
from random import randint

```

```python
df = pd.read_csv('cleaned_tree_2.csv')
```

```python
multiple_rows = []
for i, row in df.iterrows():
    multiple_rows.append(pd.DataFrame([row.values.tolist()] * row.number,
                                      columns=row.index.values))
multiple_rows = pd.concat(multiple_rows)
```

```python
multiple_rows.reset_index(drop=True)
```

```python
multiple_rows['latr'] = multiple_rows.Latitude + (
    np.random.uniform(0.001, 0.009, size=len(multiple_rows)) * np.random.choice([1, -1], size=len(multiple_rows)))
```

```python
multiple_rows.number = [1 for n in multiple_rows.number] 
```

```python
multiple_rows.Latitude = [str(d)[:-4]+ str(randint(100, 9999)) for d in multiple_rows.Latitude]
multiple_rows.Longitude = [str(d)[:-4]+ str(randint(100, 9999)) for d in multiple_rows.Longitude]
```

```python
#Це спосіб заповнити пусті клітинки замість колишніх об'єднаних
sample = df['tree_cut_requester']

temp_list = []
temp = 0
for s in sample:
    if type(s) == str: 
        temp = s
        temp_list.append(s)
    else:
        s = temp
        temp_list.append(s)
        
temp_list   
```

```python
df['tree_cut_requester'] = temp_list
       
```

```python
df['was_cut'] = [True if type(d) == str else False for d in df['cutting_act']]
#df['was_cut'] = [1 if d == True else 0 for d in df['was_cut']]
```

```python
df['date_act_published'] = [d.split(" від ")[1][3:].str for d in df.act_number]
```

```python
df['date_number'] = [d.split(' від ')[0] if type(d) == str else [d,d][0] for d in df['date']]
df['date_UTF'] = [d.split(' від ')[1] if type(d) == str else [d,d][1] for d in df['date']]
```

```python
df['tree_andress'] = ['м. Кропивницький '+d for d in df['tree_andress']]
```

```python
df.cutting_act = df.date_UTF
df.head()
```
